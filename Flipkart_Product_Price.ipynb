```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
item = input('Enter the item name: ')
```

```python
page_name = []
page_price = []
page_rating = []
page_org_price = []

page_num = input('Enter the number of pages: ')
for i in range(1,int(page_num)+1):
  url = 'https://www.flipkart.com/search?q='+item+'&page='+str(i)

  req = requests.get(url)
  content = BeautifulSoup(req.content, 'html.parser')
  
  name = content.find_all('div',{'class':'_4rR01T'})
  price = content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
  rating = content.find_all('div',{'class':'_3LWZlK'})
  org_price = content.find_all('div',{'class':'_3I9_wc _27UcVY'})

  for i in name:
    page_name.append(i.text)
  for i in price:
    page_price.append(i.text)
  for i in rating:
    page_rating.append(i.text)
  for i in org_price:
    page_org_price.append(i.text)
```

```python
data = {'Name':page_name, 'Price':page_price, 'Rating':page_rating, 'Org_Price':page_org_price}

df = pd.DataFrame.from_dict(data, orient='index').transpose()

df = df.dropna()
```

```python
df['Price'] = df['Price'].replace({'₹':'',',':'','â‚¹':''}, regex=True)
df['Org_Price'] = df['Org_Price'].replace({'₹':'',',':'','â‚¹':''}, regex=True)

df['Price'] = pd.to_numeric(df['Price'])
df['Org_Price'] = pd.to_numeric(df['Org_Price'])
```

```python
df.sort_values(by='Price', ascending=False, ignore_index=True)
```
